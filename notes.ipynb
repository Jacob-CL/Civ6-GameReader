# NOTES
### AI_CityBuild
- The decision each city makes in terms of construction. When it SUBMITS to make a unit, building or district as well as when it's COMPLETED or even PURCHASED.
- Can use to alert on new units, buildings and districts
- (maybe city was nuked?)

### City_BuildQueue
- Seems to contain only when a unit, building or district is added to a city queue
- No player numbers, just city names
- Doesn't seem to record cancelled production orders

### AStar_GC
- X, Y co-ordinates of each unit - from x, from y, to x, to y.
- Can use in tracking unit positions. -9999 X and Y means unit hasn't moved at all, or yet.

### AStar_APP
- Like the above but only for player 0's units and A LOT of repetive logs

### AI_Behaviour_Trees
- Huge .csv that I think logs each decision the AI makes and why?
